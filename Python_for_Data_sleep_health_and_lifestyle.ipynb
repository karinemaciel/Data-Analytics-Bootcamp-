## Exploratory Analysis

```python
# install the library for importing data.
import pandas as pd
```

```python
# Importing DataFrame from the challenge and visualizing the table
df = pd.read_csv('saude_do_sono_estilo_vida.csv')
df.head()
```

#### 1.
When visualizing the dataframe, you noticed that it would be better to change the names of some columns. Change 'ID' to 'Identifier', correct the name of the column indicating blood pressure, change the column 'Occupation' to 'Profession', and the column 'BMI Category' is partially in English, since it's a dataframe in Portuguese, replace it with 'IMC Category'

```python
df = df.rename(columns={'ID': 'Identificador',
                        'Pressão sanguíneaaaa': 'Pressão sanguínea',
                        'Ocupação': 'Profissão',
                        'Categoria BMI': 'Categoria IMC'})

df.columns
```

#### 2.
What is the mean, mode, and median of sleep hours for each profession?


```python
prof = df.groupby('Profissão')
media_prof = prof['Duração do sono'].mean()
moda_prof = prof['Duração do sono'].apply(lambda x: pd.Series.mode(x)[0])
mediana_prof = prof['Duração do sono'].median()

exercise_2 = pd.DataFrame({
    'Média': media_prof,
    'Moda': moda_prof,
    'Mediana': mediana_prof})

exercise_2
```

#### 3.
What percentage of people working in software engineering are obese?

```python
es = df[df['Profissão'] == 'Eng. de Software']
total_engenheiros = len(es)
obesos_engenheiros = len(es[es['Categoria IMC'] == 'Obesidade'])
porcentagem_obesos = (obesos_engenheiros / total_engenheiros) * 100
print('The percentage of people working in software engineering is', porcentagem_obesos,'%')
```

#### 4.
According to the data, does practicing law or being a sales representative make you sleep less?


```python
adv_rep = df['Profissão'].isin(['Advogado(a)', 'Representante de Vendas'])
df_adv_rep = df[adv_rep]
media_adv_rep = df_adv_rep['Duração do sono'].mean()
media_geral = df['Duração do sono'].mean()

print('Average sleep duration of lawyers and sales representative:', media_adv_rep)
print("Overall average sleep duration:", media_geral)
if media_adv_rep < media_geral:
  print('According to the data, practicing law or being a sales representative does make you sleep less.')
else:
  print('According to the data, practicing law or being a sales representative does not make you sleep less.')

```

#### 5.
Among those who studied nursing and those who studied medicine, who has fewer hours of sleep?

```python
filtro_enf_med = df['Profissão'].isin(['Enfermeiro(a)', 'Médico(a)'])
df_enf_med = df[filtro_enf_med]

media_enf = df_enf_med[df_enf_med['Profissão'] == 'Enfermeiro(a)']['Duração do sono'].mean()
media_med = df_enf_med[df_enf_med['Profissão'] == 'Médico(a)']['Duração do sono'].mean()

print("Average sleep duration for nursing:", media_enf, "and average sleep duration for medicine:", media_med)

if media_enf < media_med:
    print('Nursing professionals have fewer hours of sleep.')
else:
    print('Medical professionals have fewer hours of sleep.')
```

#### 8.
Who has the highest average blood pressure, men or women?

```python
df[['Pressão Sistólica', 'Pressão Diastólica']] = df['Pressão sanguínea'].str.split('/', expand=True)
df['Pressão Sistólica'] = pd.to_numeric(df['Pressão Sistólica'], errors='coerce')
df['Pressão Diastólica'] = pd.to_numeric(df['Pressão Diastólica'], errors='coerce')

media_pressao_homens_sistolica = df[df['Gênero'] == 'Homem']['Pressão Sistólica'].mean()
media_pressao_homens_diastolica = df[df['Gênero'] == 'Homem']['Pressão Diastólica'].mean()

media_pressao_mulheres_sistolica = df[df['Gênero'] == 'Mulher']['Pressão Sistólica'].mean()
media_pressao_mulheres_diastolica = df[df['Gênero'] == 'Mulher']['Pressão Diastólica'].mean()

media_pressao_homens = (media_pressao_homens_sistolica + media_pressao_homens_diastolica) / 2
media_pressao_mulheres = (media_pressao_mulheres_sistolica + media_pressao_mulheres_diastolica) / 2

if media_pressao_homens > media_pressao_mulheres:
    print('Men have a higher average blood pressure, with the value of:', media_pressao_homens)
else:
    print('Women have a higher average blood pressure, with the value of:', media_pressao_mulheres)
```

#### 9.
Is it predominant among participants to sleep 8 hours per day?

```python
moda_duracao_sono = df['Duração do sono'].mode()

# Verifica se a moda é igual a 8 horas
if (moda_duracao_sono >= 8).any():
    print("Yes, the predominant sleep duration among participants is 8 hours per day or more.")
else:
    print("No, the predominant sleep duration among participants is less than 8 hours per day.")
```

#### 10.
Do people with heart rates above 70 take more steps than people with heart rates less than or equal to 70?

```python
media_passos_acima_70 = df[df['Frequência cardíaca'] > 70]['Passos diários'].mean()
media_passos_70_ou_menos = df[df['Frequência cardíaca'] <= 70]['Passos diários'].mean()

if media_passos_acima_70 > media_passos_70_ou_menos:
  print('Yes, people with heart rates above 70 take more steps than people with heart rates equal to or less than 70.')
else:
    print('No, people with heart rates above 70 take fewer steps than people with heart rates equal to or less than 70.')
```
