####1. Qual dos cenários tem a maior taxa de conversão?


```python
#Total de visitantes para o cenário A e B (grupo A também pode ser chamado de grupo controle e grupo B de grupo teste /tratamento)
visitantes_a = df[df['Versão_Página'] == 'A']['Visitante_ID'].count()
print(f'Visitantes da página A:{visitantes_a}')
visitantes_b = df[df['Versão_Página'] == 'B']['Visitante_ID'].count()
print(f'Visitantes da página B: {visitantes_b}')

#Taxa de conversão A = número de visitas com conversão = 1 /número de visitas
#Taxa de conversão B = número de visitas com conversão = 1 /número de visitas
taxa_com_a = df[(df['Versão_Página'] == 'A') & (df['Conversões'] == 1)]['Visitante_ID'].count()
taxa_conv_a = taxa_com_a / visitantes_a
print(f'Taxa de conversão para página A: {taxa_conv_a:.2%}')
taxa_com_b = df[(df['Versão_Página'] == 'B') & (df['Conversões'] == 1)]['Visitante_ID'].count()
taxa_conv_b = taxa_com_b / visitantes_b
print(f'Taxa de conversão para página B: {taxa_conv_b:.2%}')

if taxa_conv_a > taxa_conv_b:
  print('Conclusão: o cenário A tem a maior taxa de conversão')
else:
  print('Conclusão: o cenário B tem a maior taxa de conversão')

#Taxa de conversão A = número de visualização com conversão = 1 /número de visualizações
#Taxa de conversão B = número de visualização com conversão = 1 /número de visualizações

```

####2. Calcule qual o tamanho da amostra necessária para o desenvolvimento de um teste A/B, seguindo os seguintes critérios:

 a. O cenário A, da base, como o inicial, que funciona hoje.

 b. Considere 95% de confiança de que o efeito na conversão não foi
 aleatório com um nível de significância de 5% (alpha = 0,05).

 c. Também considere 80% de certeza conseguir capturar o efeito
 da nova abordagem.

 d. O aumento para 10% de conversão

```python
# critérios
alpha = 0.05
beta = 0.20
delta = 0.10  # Diferença mínima detectável (aumento para 10% de conversão)
confianca = 0.95

# Calcular Z para um nível de confiança de 95% (alpha/2 = 0.025)
Z_alpha_2 = norm.ppf(1 - (1 - confianca) / 2)
print(Z_alpha_2 )
# Calcular Z para um poder estatístico de 80%
Z_beta = norm.ppf(1 - beta)
print(Z_beta)
# Estimativa da variância (usando a taxa de conversão de 10%)
variância = 0.10

# Calcular o tamanho da amostra necessário
tamanho_amostra = ((Z_alpha_2 + Z_beta) ** 2 * variância) / delta ** 2

# Imprimir o tamanho da amostra necessário
print("Tamanho da amostra necessário:", tamanho_amostra.round(), "dados.")
```

Considere o nível de significância a 6%.

```python
amostra_5 = dfp.loc[[690, 894, 67, 201, 364, 19, 60, 319, 588, 643, 855, 623, 530, 174, 105, 693, 6, 462, 973, 607, 811, 346, 354, 966, 943, 372]]['Pressao_Arterial']
pressao_arterial_saud = dfp[dfp['Estado_Saude'] == 0]['Pressao_Arterial']
pressao_arterial_n_saud = dfp[dfp['Estado_Saude'] == 1]['Pressao_Arterial']

media_amostra_5 = amostra_5.mean()

media_pressao_arterial_saud = pressao_arterial_saud.mean()
desvio_padrao_pressao_arterial_saud = pressao_arterial_saud.std()

# Tamanho das amostras
n_amostra_5 = len(amostra_5)
n_pressao_arterial_saud = len(pressao_arterial_saud)

# Calculando a estatística Z
Z = ((media_amostra_5 - media_pressao_arterial_saud) / (desvio_padrao_pressao_arterial_saud / (n_amostra_5 ** 0.5)))

# Calculando o p-valor usando a distribuição normal padrão
p_value = norm.cdf(Z)

# Definindo o nível de significância
nivel_significancia = 0.06

# Comparando o p-valor com o nível de significância
if p_value < nivel_significancia:
    print("Rejeitamos a hipótese nula. A média da pressão arterial para pacientes com condições de saúde adicionais é diferente da média da pressão arterial para pacientes sem essas condições.")
else:
    print("Não há evidências suficientes para rejeitar a hipótese nula. A média da pressão arterial para pacientes com condições de saúde adicionais é igual à média da pressão arterial para pacientes sem essas condições.")
```

6. Existe uma diferença significativa na pressão arterial média entre diferentes grupos étnicos nesta população? (Teste ANOVA, alpha é 5%)

a. Hipótese Nula (H0): A pressão arterial média é a mesma em todos os grupos étnicos.

b. Hipótese Alternativa (H1): Há uma diferença significativa na pressão arterial média entre pelo menos dois grupos étnicos nesta população.

```python
# Separando os grupos étnicos
grupo_etnico_1 = dfp[dfp['Etnia'] == 1]['Pressao_Arterial']
grupo_etnico_2 = dfp[dfp['Etnia'] == 2]['Pressao_Arterial']
grupo_etnico_3 = dfp[dfp['Etnia'] == 3]['Pressao_Arterial']

# Teste ANOVA
f_statistic, p_value = f_oneway(grupo_etnico_1, grupo_etnico_2, grupo_etnico_3)

# Definindo o nível de significância
alpha = 0.05

# Comparando o p-valor com o nível de significância
if p_value < alpha:
    print("Rejeitamos a hipótese nula. Existe uma diferença significativa na pressão arterial média entre diferentes grupos étnicos nesta população.")
else:
    print("Não há evidências suficientes para rejeitar a hipótese nula. Não existe diferença significativa na pressão arterial média entre diferentes grupos étnicos nesta população.")
```

 9. Qual é o intervalo de confiança para a média da pressão arterial entre
 os pacientes com condições de saúde adicionais? (nível de confiança 95%)

```python
# Amostra de pressão arterial para pacientes com condições de saúde adicionais
pressao_arterial_n_saud = dfp[dfp['Estado_Saude'] == 1]['Pressao_Arterial']

media = pressao_arterial_n_saud.mean()
desvio_padrao = pressao_arterial_n_saud.std()

n = len(pressao_arterial_n_saud)
nivel_confianca = 0.95
# Graus de liberdade para a distribuição t de Student
graus_liberdade = n - 1

# Calculando o valor crítico da distribuição t de Student
valor_critico = stats.t.ppf((1 + nivel_confianca) / 2, graus_liberdade)

# Calculando o erro padrão da média
erro_padrao_media = desvio_padrao / np.sqrt(n)

# Calculando o intervalo de confiança
intervalo_confianca = (media - valor_critico * erro_padrao_media, media + valor_critico * erro_padrao_media)

print("Intervalo de Confiança para a Média da Pressão Arterial entre Pacientes com Condições de Saúde Adicionais (95% de confiança):", intervalo_confianca)
```
